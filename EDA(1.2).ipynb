```python
from sklearn.linear_model import  BayesianRidge,PoissonRegressor,PassiveAggressiveRegressor,ARDRegression,Ridge,Lasso,ElasticNet,LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
```

```python
x=df.drop(columns=['Selling_Price'],axis=1)
Y=df[['Selling_Price']]
```

```python
x
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,Y_train,Y_test=train_test_split(x,Y,test_size=0.30,random_state=21)
```

```python
import numpy as np
def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    r2_square = r2_score(true, predicted)
    return mae, r2_square,mse
```

```python
models={
    'BayesianRidge':BayesianRidge(),
    'PoissonRegressor':PoissonRegressor(),
    'PassiveAggressiveRegressor':PassiveAggressiveRegressor(),
    'ARDRegression':ARDRegression(),
    'Ridge':Ridge(),
    'Lasso':Lasso(),
    'ElasticNet':ElasticNet(),
    'LinearRegression':LinearRegression()
}
trained_model_list=[]
model_list=[]
r2_list=[]

for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(x_train,Y_train)

    #Make Predictions
    y_pred=model.predict(x_test)

    mae, r2_square,mse=evaluate_model(Y_test,y_pred)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])
    

    print('Model Training Performance')
   
    print("MAE:",mae)
    print("R2 score",r2_square)

    r2_list.append(r2_square)
    
    print('-'*30)
    print('\n')
```

```python
linear_regression=Lasso()
linear_regression.fit(x_train,Y_train)
```

```python
import pickle
model_pkl_file="modellasso.pkl"

with open(model_pkl_file,'wb') as file:
    pickle.dump(linear_regression,file)
```
